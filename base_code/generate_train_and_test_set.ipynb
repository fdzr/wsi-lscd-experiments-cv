```python
from pprint import pprint

import pandas as pd
from sklearn.model_selection import KFold, RepeatedKFold
```

```python
random_state = 67678
```

```python
rkf = RepeatedKFold(n_splits=5, n_repeats=5, random_state=random_state)
```

```python
X = pd.read_csv("../test_data_es.csv", sep="\t").word.tolist()
X
```

```python
for train, test in rkf.split(X):

    print("train: ", train)
    print("test: ", test)
    print()
```

```python
for train, test in rkf.split(X):
    training_set = list(map(X.__getitem__, train))
    testing_set = list(map(X.__getitem__, test))
    print(training_set)
    print(testing_set)
    print()
    
```

```python
kf = KFold(n_splits=6, shuffle=True, random_state=456123)
```

```python
for train, test in kf.split(X):
    train_set = list(map(X.__getitem__, train))
    test_set = list(map(X.__getitem__, test))

    print(train_set)
    print(test_set)
    print()
```
