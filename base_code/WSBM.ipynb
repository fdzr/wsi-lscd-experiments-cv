```python
from datetime import datetime

from common_new_functionalities import (
    load_dwug_sense,
    grid_search_without_nclusters,
)
import clustering
import numpy as np
```

```python
def get_clusters(adj_matrix, hyperparameters, seed=0):
    graph = clustering._adjacency_matrix_to_nxgraph(
        adj_matrix,
        use_disconnected_edges=hyperparameters["use_disconnected_edges"],
    )
    clusters = clustering.wsbm_clustering(graph, **hyperparameters)
    return clustering._convert_graph_cluster_list_set_to_list(graph, clusters)
```

```python
model_hyperparameter_combinations = []

for distribution in [
    "discrete-geometric",
    "discrete-poisson",
    "discrete-binomial",
    # "real-normal",
    # "real-exponential"
]:
    model_hyperparameter_combinations.append({"distribution": distribution})
```

```python
method = "wsbm"
is_cross_validation=True
path_to_experiments = {
    True: "../cv-experiments/wsbm-llama/" + "{kfold}_fold",
    False: "../no-cv-experiments-wsbm/" + f"{method}/" + "{kfold}"
}
```

```python
# logging.basicConfig(
#     filename="logs/wsbm/logs.txt",
#     filemode="a+",
#     format="%(asctime)s : %(message)s",
#     datefmt="%y-%m-%d %I:%M:%S %p",
#     level=logging.INFO
# )
```

```python
start_time = datetime.now()

grid_search_without_nclusters(
    load_dwug_sense,
    get_clusters,
    model_hyperparameter_combinations,
    include_binarize=False,
    method=method,
    logger_message=None,
    cache=path_to_experiments[is_cross_validation],
    run_experiments=1,
    dataset="dwug_data_annotated_only",
    is_cross_validation=is_cross_validation
)

print(f"Elapsed time: {datetime.now() - start_time}")
```
