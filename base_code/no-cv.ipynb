```python
input = "../input/{llm}/{dataset}/fs"
```

```python
gold_data_spanish_compare = gold_data_spanish.set_index("word")["COMPARE"].to_dict()
```

```python
for dataset in ["dwug_en"]:
    print(f"{dataset}:")

    for model in ["llama3.1-8B", "mixtral-8xtb-v0.1"]:
        print(f"  {model}:", end=" ")

        p = Path(input.format(llm=model, dataset=dataset))
        v1, v2 = [], []

        assert p.exists() is True

        data = load_data(p)
        if dataset == "dwug_en":
            data = data.apply(lambda row: build_id(row), axis=1)

        mask_earlier = (data["identifier1"].str.split("_").str[0] == "old") & (
            data["identifier2"].str.split("_").str[0] == "old"
        )
        mask_later = (data["identifier1"].str.split("_").str[0] == "new") & (
            data["identifier2"].str.split("_").str[0] == "new"
        )
        mask_compare = (~mask_earlier & ~mask_later) | (~mask_later & ~mask_earlier)

        data_compare = data[mask_compare]
        data_earlier = data[mask_earlier]
        data_later = data[mask_later]

        pred_earlier = data_earlier.groupby("word")["score"].mean().reset_index()
        pred_later = data_later.groupby("word")["score"].mean().reset_index()
        pred_compare = data_compare.groupby("word")["score"].mean().reset_index()

        compare_earlier = pred_compare.merge(
            pred_earlier, on="word", suffixes=("_compare", "_earlier")
        )
        compare_earlier_later = compare_earlier.merge(pred_later, on="word")
        compare_earlier_later.rename(columns={"score": "score_later"}, inplace=True)

        compare_earlier_later["score"] = compare_earlier_later["score_compare"] / (
            compare_earlier_later["score_later"]
            - compare_earlier_later["score_earlier"]
        )
        # compare_earlier_later["score"] = compare_earlier_later[
        #     "score_compare"
        # ] / compare_earlier_later[["score_earlier", "score_later"]].max(axis=1)

        pred = compare_earlier_later.set_index("word")["score"].to_dict()

        for tw in target_words:
            v1.append(gold_data_change_graded[tw])
            v2.append(-1 * pred[tw])

        spr, _ = spearmanr(v1, v2)
        print(spr)
```
